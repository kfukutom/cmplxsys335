```python
# Question 2:
# args: limit to instantiate the maximum value
def prime(lim):
  if lim < 2:
    return []

  primes = [2]
  for i in range(3, lim+1):
    status = True
    for j in primes:
      if (i % j == 0):
        # base case, divisible means /= prime
        break
      if (j * j > i):
        # check up to sqrt(i)
        break
    if status == True:
      primes.append(i)

  return primes
```

```python
import time

start = time.time()
x = prime(10000)
end = time.time()

print(f"The list up primes up to 10000 is {x}\n")
print(f"Length: {len(x)} & Time {round(end-start, 3)} seconds.")
```

# 2 Stock market

```python
from csv import reader
from datetime import datetime
from pylab import plot, show, xlim

x, y = [], []

with open('dow.csv', 'r') as file:
    r = reader(file)

    # date = datetime.strptime(date_str, "%Y-%m-%d")
    for data in r:
      d = datetime.strptime(data[0], "%Y-%m-%d")
      x.append(d)
      y.append(float(data[1]))

file.close()

plot(x,y,'b')
show()
```

Question 2)

```python
with open('dow.csv', 'r') as file:
    r = reader(file)
    final = [] # Filtering list with ONLY January Dates.
    x, y, z = [], [], []

    for data in r:
      d = datetime.strptime(data[0], "%Y-%m-%d")
      if (d.month == 1):
        final.append(d)
        z.append(float(data[1]))

# Iterate by years:
iter = {}

for i, data in enumerate(final):
  d = final[i]
  year = d.year

  if d.year not in iter:
    iter[year] = True
    x.append(d)
    y.append(z[i])
```

```python
pct_change = [0] # 0 applied to first date.
for i in range(1, len(y)):
  pct = float(((y[i] - y[i-1]) / y[i-1]) * 100)
  pct_change.append(pct)

assert len(pct_change) == len(x) and len(y)
plot(x,pct_change,'g')
show()
# Average Pct Gain
gain = sum(pct_change) / len(pct_change)
print(f"The average yearly percentage gain over all years in the data set is {round(gain, 4)}%\n")
```

```python
from networkx import Graph, draw, read_adjlist, DiGraph
from networkx.drawing.nx_pydot import pydot_layout
import networkx as nx
import matplotlib.pyplot as plt
from numpy import array
```

```python
G = DiGraph()
file = open('attweb.txt', 'r')
for line in file:
  data = array(line.split(), int)
  i = data[0]
  for j in data[1:]:
    G.add_edge(i, j)

file.close()

p = pydot_layout(G)
draw(G, node_size=80, pos=p)
#draw(G, node_size=100, node_color='red') # Optional Arguments
```

```python
#CSV Function
x, y = [], []

with open('students.csv', 'r') as file:
    r = reader(file)

    for student in r:
      x.append(student[0])
      y.append(student[1])
```

```python
G = Graph()

file = open('dating.txt', 'r')
for line in file:
  data = array(line.split(), int)
  i = data[0]
  for j in data[1:]:
    G.add_edge(i, j)
file.close()

color_map = []
# Coloring
for node in G:
  var = node-1
  if y[var] == 'F':
    color_map.append('pink')
  elif y[var] == 'M':
    color_map.append('blue')

# Node Size, random scale factor
degrees = [9.75 * G.degree[i] for i in G]
#print(G.degree[3])

p = pydot_layout(G)
nx.draw_networkx_nodes(G, p, node_size = degrees, node_color = color_map)
nx.draw_networkx_edges(G, p, alpha=0.85, edge_color = 'black')
plt.show()
```
