```python
from numpy import loadtxt, sum, zeros, empty
import networkx as nx
from networkx.drawing.nx_pydot import pydot_layout
from pylab import plot
```

```python
# Efficient Way to Run the Drawing
G = nx.read_adjlist('highschool (1).txt', create_using=nx.DiGraph, nodetype=int)
n = G.number_of_nodes()
histogram = zeros(n, int)
# Each element of the histogram is accessed and incremented upon iterating through the G
for i in G:
  d = (G.degree[i])
  histogram[d] += 1
```

```python
plot(histogram)
```

```python
# Degree: The number of connections a node has. It's a very simple property,
# but it's quite literally the most important concept to grasp in network science.
```

**Centrality Measure**
- Degree as some centrality measure, hence stating that, a node with more edges is deeemed more influential and important in a network. Example: a person you know who has a lot of connections. Successful business people know a lot of people. A very simple measure, but very useful to identify certain individuals, nodes that are '*important*'.

**Out Degree/In Degree**
- Out degree could be the number of people who say you're friends with. In degree, other people call/consider you as a friend. Another example, consider a web network, a page has edges that link to other pages. We can also calculate this degree using an adjacency matrix. One, iterate through a row within a two-dimensional array, and find the number of 1's are present, which indicate a connection from that node (row).

```python
#for i in range(len(A)):
  #print(sum(A[i]))

A = loadtxt('karate.mat', int)
d = (sum(A, axis=1))
G = nx.from_numpy_array(A)
size = []
for i in G:
  size.append(40 * d[i])
print(size)

#p = pydot_layout(G)
nx.draw(G, with_labels=True, node_color=size, node_size=size)
```

```python
data = loadtxt('webin.txt', int)
print(data) # <- initially, the data is arranged as two columns of arrays.
```

```python
# step 2, split into readable array
from pylab import plot
from pylab import xscale, yscale

x = data[:,0]
y = data[:,1]
N = len(y)
yc = empty(N, int)
for k in range(N):
  yc[k] = sum(y[k:])
# A cumulative distribution plot. Downside, it's harder to interpret.

plot(x, yc) # Previous we used plot(x, y)
xscale("log")
yscale("log")
# this is a huge network given it's result. let's apply the log-log plot
```
